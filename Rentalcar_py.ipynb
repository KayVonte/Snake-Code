```python
# NAME: rentalcar.py
# IMPORT STATEMENTS
# (none [math removed])

# FUNCTION DEFINITIONS
# 1. Assign car classification codes (B, S, L).
def car_classification():
  while True:
    code = input("Enter car classification code (B = Budget, S = Standard, L = Luxury): ").upper()
    if code in ['B', 'S', 'L']:
      return code
    else:
      print("Invalid classification. Please enter B, S, or L.")

# - Define classification name (code).
def classification_name(code):
  names = {
    'B': "Budget",
    'S': "Standard",
    'L': "Luxury"
  }
  return names[code]

# - Define daily rate.
def daily_rate(code):
  rates = {
      'B': 40,
      'S': 60,
      'L': 80
  }
  return rates[code]

# 2. Define start and end odometer (mileage).
def mileage():
  while True:
    try:
      start_mileage = float(input("Enter starting odometer reading: "))
      end_mileage = float(input("Enter ending odometer reading: "))
      if end_mileage >= start_mileage:
        return end_mileage - start_mileage
      else:
        print("Invalid input. Ending mileage must be greater than or equal to starting mileage.")
    except ValueError:
      print("Invalid input. Please enter a valid number.")

# - Define extra miles(total_miles)
def extra_miles(total_miles):
  free_miles = 100
  if total_miles > free_miles:
    return total_miles - free_miles
  return 0

# 3. Define days rented(days)
def days_rented():
  while True:
    days_str = input("Enter number of days rented: ")
    try:
      days = int(days_str)
      if days > 0:
        return days
      else:
        print("Insufficient number of days. Please enter a positive number.")
    except ValueError:
      print("Invalid input. Please enter a number.")

# 4. Define total bill cost (mileage, classification).
def calculate_total_bill(days, rate, chargeable_miles):
  mileage_rate = 0.35
  rental_cost = days * rate
  mileage_cost = chargeable_miles * mileage_rate
  total = rental_cost + mileage_cost
  return total

# PROGRAM LOGIC
def main():
  print("=== Rental Car Billing System ===")

  # 1. Input for car classification code.
  code = car_classification()
  car_type = classification_name(code)
  daily_rental_rate = daily_rate(code)

  # 2. Input for start and end odometer.
  total_miles_driven = mileage()
  chargeable_miles = extra_miles(total_miles_driven)

  # 3. Input for days rented.
  days = days_rented()

  # 4. Call and print the total bill cost.
  total_bill = calculate_total_bill(days, daily_rental_rate, chargeable_miles)

  print("\n----- RENTAL CAR SUMMARY  -----")
  print("Car Classification:", car_type)
  print("Total Miles Driven:", total_miles_driven)
  print("Chargeable Miles:", chargeable_miles)
  print("Finalized Bill: $", format(total_bill, '.2f'))

if __name__ == "__main__":
  main()
```
